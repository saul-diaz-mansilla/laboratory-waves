# Laboratory Waves: Results Analysis

This notebook provides a detailed overview of the conclusions extracted from the various analyses and simulations of the discrete LC transmission line, as discussed in the project's README.

## 04 Base Model Comparison
![Base Model Comparison](../figures/04_base_model_comparison.png)

**Conclusion:** The initial comparison between the ideal theoretical model (which assumes zero tolerance on L and C, and no inductor resistance) and the experimental data highlights significant discrepancies. The ideal low-pass filter behavior fails to account for real-world signal standing waves and resonances, as different values of L and C from section to section modify the internal impedances, giving rise to internal reflections.

## 05 Dispersion Relation
![Dispersion Relation](../figures/05_dispersion_relation.png)

**Conclusion:** The experimental dispersion relation falls neatly within the shaded region created through upper and lower bounds for Monte-Carlo simulations, proving that model and experiment are consistent. Furthermore, it shows that both dispersion relations follow closely a sinusoidal relation given by $f(k)=f_c \sin{\frac{k}{2l_0}}$, which corresponds to an ideal transmission line, so both ideal and dispersive Monte-Carlo models produce extremely similar dispersion relations.

## 06 Scaling Comparison
![Scaling Comparison](../figures/06_scaling_comparison.png)

**Conclusion:** Incorporating dynamic, frequency-dependent series resistance for the inductors is crucial. By applying a frequency scaling exponent, the simulated high-frequency dispersion and signal attenuation align much more closely with the experimental data, successfully correcting divergences seen in non-scaled basic models.

## 07 Trend Comparison
![Trend Comparison](../figures/07_trend_comparison.png)

**Conclusion:** By comparing linear trends of the transfer function against Monte Carlo distributions, we see a that experimental and simulated trends match in the mean, with a spread inherent to the Monte-Carlo methods used. We see that the exponent chosen for the frequency scaling for the inductor resistance matches the trend successfully.

## 09 Inference Parameters (1D and 2D)
![1D Inference](../figures/09_infer_parameters_1d.png)

**Conclusion:** We now attempt to extract information of the parameters of the experimental transmission line using a convolutional CNN. The most robust and and efficient approach is to use gaussian inputs to calculate the transfer function. As we do not have enough data to validate this apporach on all measurable nodes, we train a simplified version of the CNN using just the last node. We see that this approach fails as the model does not infer the parameters correctly.

## 10 Inference Simulated 2D
![Inference Validation](../figures/10_infer_simulated_2d.png)
![Inference Validation](../figures/10_infer_simulated_2d_magnitude.png)

**Conclusion:** The overarching validation of the inverse problem. Comparing the transfer function reconstructed from the CNN's physical parameter predictions against the ground truth confirms excellent agreement. The NN has correctly learnt the physics of the problem and has managed to solve the inverse problem to great accuracy for 50 epochs.

This agreement happens both for matched (top) and unmatched (bottom) simulations, and it is observed that for the bottom figure, which just trains on the magnitudes of the trasnfer function, the convolution kernels are able to learn more accurately how the resonances are related to the internal parameters of the transmission line.

More experimental data is needed, either from gaussians or pulses (whose first pole lies beyond the cutoff frequency), as transfer functions do not depend on the input signal, to fully validate the neutral network approach.
